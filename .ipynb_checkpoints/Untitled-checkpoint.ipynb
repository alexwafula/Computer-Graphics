### Generate the email addresses of all the students in the format below

```python
for i in data3B.index:
    special_char=['`','"',"'"]
    
    name=data3B.iloc[i,2]
    first_name=name.split(',')[0]
    last_name=name.split()[-1]
    #print(first_name,last_name)
    email=str(first_name[0]).lower()+last_name.lower()+'@gmail.com'
    for character in special_char:
        email=email.replace(character,'')
    print(email)
    #Check if the email already exists in the dataset
    if email not in data3B.iloc[0:,4].unique():
        data3B.iloc[i,4]=email

for i in data3C.index:
    special_char=['`','"',"'"]
    
    name=data3C.iloc[i,2]
    first_name=name.split(',')[0]
    last_name=name.split()[-1]
    #print(first_name,last_name)
    email=str(first_name[0]).lower()+last_name.lower()+'@gmail.com'
    for character in special_char:
        email=email.replace(character,'')
    print(email)
    #Check if the email already exists in the dataset
    if email not in data3B.iloc[0:,4].unique():
        data3C.iloc[i,4]=email
```

```python
import json

json_data2=[]
for i in new_data.index:
    student_data={
        'id':str(i),
        'student_number':str(new_data.iloc[i,1]),
        'additional_details':[
            {
                'dob':str(new_data.iloc[i,3]),
                'gender':str(new_data.iloc[i,5]).lower(),
                'special_character':'yes',
                'name_similar':'no'
            }
        ]
    }
    
    json_data2.append(student_data)

#Writing the new json data to a new file
file=open('json_file2.json','w')
data=json.dumps(json_data2)
file.write(data)
file.close()
```

```python
from googleapiclient.discovery import build
from google.oauth2 import service_account
import os
```

```python
api_url=['https://www.googleapis.com/auth/drive']
service_account_file='keys.json'
parent_folder_id='15hQAbzVfDTgCB3-rXOW7DAVNFiyA4ENL'

def auth():
    creds=service_account.Credentials.from_service_account_file(service_account_file)
    return creds
auth()

def upload(path):
    creds=auth()
    service=build('drive','v3',credentials=creds)
    filename=path.split('.')[0]
    file_metadata= {
        'name':filename,
        'parents':[parent_folder_id]
    }
    file=service.files().create(
        body=file_metadata,
        media_body=path
    ).execute()
```

```python
#All files in directory
files=os.listdir('.')
for file in files:
    try:
        upload(file)
    except:
        print('Uploading file:'+(file)+' failed')
```
